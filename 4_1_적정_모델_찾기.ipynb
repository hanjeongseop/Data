### 1 Scaling : 숫자데이터의 서로 다른 차원(scale)을 맞춰주는 작업

```python
import pandas as pd
```

```python
# 파일 읽기 및 내용 확인하기

github_csv_url = 'https://raw.githubusercontent.com/hanjeongseop/Data/main'

df1 = pd.read_csv(github_csv_url + '/Contract.csv')

df1.info()
```

```python
# 신용등급, 연체건수, 나이, 계약금액, 계약개월
df2 = df1[['Credit_Rank', 'Overdue_count', 'Age', 'Amount_Month', 'Term']]
```

```python
df2.describe()
```

```python
# 변수 마다 독립적인 스케일을 가지고 있다.
# 따라서 스케일올 맞춰줘야 한다.

from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import RobustScaler
```

```python
# 각각의 데이터를 확인해 스케일을 변환

scaler = StandardScaler()
scaler.fit(df2)
```

```python
pd.DataFrame(scaler.transform(df2)).describe()
```

# 평균이 0으로 표준편차는 1을 기준으로 변환되었다.

# 실제 실무에서는 이상치가 존재할 수 있기때문에 RobustScaler을 많이 사용한다.

```python
scaler = RobustScaler()
scaler.fit(df2)
```

```python
pd.DataFrame(scaler.transform(df2)).describe()
```

# 특정값이 많이 존재하는 경우에는 0으로 나타날 수 있다.
# 사전에 데이터 정제 작업을 통해 해결할 수 있다.

```python
# MinMaxScalersms 분류모델을 이용하는 경우 사용한다.

scaler = MinMaxScaler()
scaler.fit(df2)
```

```python
pd.DataFrame(scaler.transform(df2)).describe()
```

# 라벨인코딩을 적용한 경우 서열화가 적용될 수 있음

```python
# One Hot 인코딩 적용
# 문자열이나 범주형 데이터를 자동으로 변환한다

pd.get_dummies(df1[['Address1']])
```

```python
df_col = pd.get_dummies(df1[['Address1']])

df3 = df1.join(df_col)
```

```python
# 계약과 연관된 주요인자 확인 (판매방식, 계약기간, 대여료, 나이, 미납횟수, 신용등급, 계약상태)

df1[['Sales_Type','Term','Amount_Month','Age','Overdue_count','Credit_Rank', 'State']]
```

```python
df3.head()
```

```python
df2 = df1[['Sales_Type','Term','Amount_Month','Age','Overdue_count','Credit_Rank', 'State']]

df2['State'].value_counts()
```

```python
df2['target'] = df2['State'].replace({'계약확정':0, '기간만료':0, '해약확정':1,'해약진행중':1})
```

```python
df2['target'].value_counts()
```

```python
df3 = df2.dropna()

df3
```

```python
Y = df3['target']
X = df3.drop(columns=['State','target'])
```

```python
pd.get_dummies(X)
```

```python
X1 = pd.get_dummies(X)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
X_train, X_test, y_train, y_test = train_test_split(X1, Y, test_size=0.3)
```

```python
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
```

```python
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)
```

```python
# 분류모델에서 모든 평가 지표를 확인할 수 있다.
from sklearn.metrics import classification_report

print(classification_report(y_train, y_train_pred))
```

```python
print(classification_report(y_test, y_test_pred))
```

## 정밀도(Precision)
 * 예측을 '참'으로 한 대상 중에 예측과 실제 값이 '참'으로 일치한 데이터의 비율

## 재현율(Recall) or 민감
 * 실제 값이 '참'인 대상 중에 예측과 실제 값이 '참'으로 일치한 데이터의 비율
  
## F1 스코어 = 정밀도와 재현율의 조화평균
 * 정밀도와 재현율이 어느 한쪽으로 치우치지 않을 때 상대적으로 높은 값을 가진다
